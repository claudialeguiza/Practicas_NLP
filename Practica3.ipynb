```python
import pandas as pd
import numpy as np
import stanza
stanza.download('spanish')
import es_core_news_lg
import spacy
```

```python
# Cargar el modelo preentrenado para español
nlp = spacy.load('es_core_news_lg')
```

```python
frases = """Cuidadoso es mi paso por la senda donde el paso del tiempo se hace eterno. Cada huella que dejo me encomienda, a un futuro que paso a paso yo gobierno.
            En la mañana, yo canto con alegría el canto del jilguero me acompaña. Mis notas se elevan, su melodía, entreteje esperanzas en la montaña.
            En la quietud de la noche, él espera, bajo la luz tenue de una vieja farola, la espera se torna eterna y sincera, mientras el viento de otoño desconsola"""
```

```python
# Procesar el texto con el modelo de spaCy
doc = nlp(frases)

# Crear una lista para almacenar las palabras, etiquetas POS y explicaciones
data = []

# Iterar sobre los tokens en el Doc y agregar los detalles a la lista de datos
for token in doc:
    data.append([token.text, token.pos_, spacy.explain(token.pos_)])

# Crear un DataFrame a partir de la lista de datos
df = pd.DataFrame(data, columns=['Palabra', 'Etiqueta POS', 'Explicación'])
```

```python
df
```

```python
sustantivos =  df[df["Etiqueta POS"] == "NOUN"]
sustantivos
```

El verbo conjugado gobierno lo toma como proper noun y la palabra canto que en el texto esta como sustantivo y verbo, lo toma ambas veces como NOUN

```python
len(sustantivos)
```

```python
verbos = df[df["Etiqueta POS"] == "VERB"]
verbos
```

El verbo conjugado gobierno lo toma como proper noun y la palabra canto que en el texto esta como sustantivo y verbo, lo toma ambas veces como NOUN. El verbo desconsolar no lo reconoce.

```python
len(verbos)
```

```python
# Texto para analizar
doc1 = nlp(frases)

# Navegamos cada oración de nuestro texto
data1 = []
for i, sent in enumerate(doc1.sentences):
      print("[Sentence {}]".format(i+1))
      for word in sent.words:
            print("{:12s}\t{:12s}\t{:6s}\t{:d}\t{:12s}".format(\
                  word.text, word.lemma, word.pos, word.head, word.deprel))
            data1.append([word.text, word.lemma, word.pos, word.head, word.deprel])
      print("")
```

```python
# Crear un DataFrame a partir de la lista de datos
df = pd.DataFrame(data1, columns=['Palabra', 'Foma Canonica', 'Etiqueta POS', \
                                  'Indice HEAD', 'Relacion HEAD'])
df
```

```python
sustantivos = df[df["Etiqueta POS"] == "NOUN"]
sustantivos
```

```python
len(sustantivos)
```
